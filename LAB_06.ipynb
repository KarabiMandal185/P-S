```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
mu= 100
sigma= 5
```

```python
data = np.random.normal(mu, sigma, 1000)
```

```python
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 10000)
pdf = norm.pdf(x, mu, sigma)
```

```python
plt.figure(figsize=(10, 6))
plt.hist(data, bins=30, density=True, alpha=0.6, color='#1f77b4', edgecolor='black', linestyle='dashed', linewidth=1.5)
```

```python
plt.plot(x, pdf,'r-',lw=2)# Add this line to plot the normal curve in red
plt.title("Normal Distribution:$\mu$=100,$\sigma$=5")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
x1=-1
x2=1
x=np.linspace(-4,4,1000)
y=stats.norm.pdf(x)
plt.plot(x,y,label='Normal Distribution')
plt.fill_between(x,y,where=(x>x1)&(x<x2),color='grey',alpha=0.5,label='P(x1<X<x2)')
plt.fill_between(x,y,where=(x<x2),color='blue',alpha=0.3,label='P(X<x2)')
plt.fill_between(x,y,where=(x>x1),color='green',alpha=0.3,label='P(x1<X)')
plt.xlabel('x')
plt.ylabel("probability density")
plt.title("Normal distribution with shaded areas")
plt.legend
plt.grid(True)
plt.show()
```
